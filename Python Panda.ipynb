```python
# Load panda library
import pandas as pd

# Load the dataset Pokemon from a csv file
df = pd.read_csv(r"C:\Users\Nezar\Downloads\Pokemon.csv")

# use the .head() method to look at the top 5 rows
df.head()

# Use the .tail method to look at the bottom 5 rows
df.tail()
```

# Reading data in panda

```python
# read headers or columns
print(df.columns)
print("--------------------------------------------------------------")

# Read single column
print('Display single column Name:')
print(df['Name'])
print("--------------------------------------------------------------")

# Read each column with slicing
print(df['Name'][5:10])
print("--------------------------------------------------------------")

# Read multiple columns
print(df[['Name','Type 1', 'Legendary']][0:5])
print("--------------------------------------------------------------")
```

```python
# Read all rows and specific range of columns between Name and Type 2
print(df.loc[:, 'Name':'Type 2'])
print("--------------------------------------------------------------")

# Read Rows using their integer index location
print(df.iloc[0:4])

# Read a specific location (cell) in the dataframe
print(df.iloc[2,1])
```

```python
# Show rows Filtered by a specific value in a column
df.loc[df['Type 1'] == 'Grass'][0:5]
```

```python
# using .describe() shows quickly a useful set of stats
df.describe()
```

```python
# Sort by Attack and in descending order
df.sort_values('Attack', ascending=False)[0:5]
```

```python
# Sort values by 2 columns, Type 1 in ascending and HP in descending order
df.sort_values(['Type 1', 'HP'], ascending=[1,0])[0:10]
```

```python
# Create a new column from the sum of other columns
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']

df.head(3)
```

```python
# Drop a specific column
df=df.drop(columns=['Total'])
df.head(3)
```

```python
# Now Add a new column named Total using the .iloc & .sum methods
df['Total'] = df.iloc[:, 4:10].sum(axis=1)
df.head(3)
df.sort_values('Total', ascending=False)[0:3]
df[['Name', 'Total']][0:5]
```

```python
# re arrange columns by using their name
# Can break code! reset from the top by loading the data again
df = df[[ 'Name', 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed', 'Generation', 'Legendary', 'Type 1', 'Type 2', '#']]
df.head()
```

```python
# re-arrange columns by using list method
# Can break code! reset from the top by loading the data again
cols = list(df.columns)
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]

df.head(3)
```

Filtering Data:

```python
# Filter by 2 conditions
df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] > 70)]

# Save the filtered results into a new data frame
new_df = df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] > 70)]
new_df.head(3)
```

```python
# notice there is a problem with the index values in re previous results, they are still the same and did not reset
# the way to fix this:
new_df1 = new_df.reset_index(drop=True)
# We use the (drop = True) to prevent the program from saving the old index values in a new column
new_df1.head(3)
```

```python
# If we want to save the dataframe with updated index in the same file without saving it somewhere else use inplace=True:
new_df.reset_index(drop=True, inplace=True)
new_df
```

```python
# filter by all the names that contain mega
df.loc[df['Name'].str.contains('Mega')][0:5]
```

```python
# filter by all the names that does NOT contain mega
df.loc[~df['Name'].str.contains('Mega')][0:5]
```

```python
import re
# filter by all Type 1 that are either fire or grass, use the flag condition that ignores capitilization
# flags=re.I <--- for capitalizatio ignoring
df.loc[df['Type 1'].str.contains('fire|grass', flags=re.I, regex=True)][0:5]
```

```python
# Filter all the names that start with pi
# The [a-z]* is used to say all that comes after doesnt matter
# the ^ before pi is to say start the line with pi
df.loc[df['Name'].str.contains('^pi[a-z]*', flags=re.I, regex=True)]
```

```python
# it turns out we didnt need the [a-z]*
df.loc[df['Name'].str.contains('^pi', flags=re.I, regex=True)]
```

Conditional Changes

```python
# Change all the fire pokemons to be named flamer
df.loc[df['Type 1'] == 'Fire', 'Type 1'] = 'Flamer'
df
```

```python
# Change the values of 2 columns based on a specific condition in another column
df.loc[df['Total'] > 500, ['Generation', 'Legendary']] = ['Test 1', 'Test 2']
df
```

Aggregate Statistics (Groupby)

```python
# Group by a tag repeate in a column
# find the mean of that tag and then sort the values by one of the columns
df.groupby(['Type 1']).mean().sort_values('HP', ascending = False)[0:5]
```

```python
# group by a tag in a column and use count
df['count'] = 1
df.groupby(['Type 1','Type 2']).count()['count']
```
